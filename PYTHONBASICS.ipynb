  ==

  is
  The is operator is used to compare the identities of two objects. It's also known as the identity operator. For example, if you compare two cats with the is operator, you'd get the answer "these are two different cats".

```python
# 14. Write a program to reverse a string using a while loop.
str = "mohit" #  string variable
print ("The original string  is : ",str)
reverse_String = ""  # Empty String
count = len(str) # Find length of a string and save in count variable
while count > 0:
    reverse_String += str[ count - 1 ] # save the value of str[count-1] in reverseString
    count = count - 1 # decrement index
print ("The reversed string using a while loop is : ",reverse_String)# reversed string
```

```python
# 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
 # 15. Write a Python program to calculate the factorial of a number provided by the user using a while loop.
def factorial(n): # Corrected indentation

    result = 1

    for i in range(1, n + 1):

        result *= i

    return result

# Example usage:

number = 5

result = factorial(number)

print(f"The factorial of {number} is {result}")
```
